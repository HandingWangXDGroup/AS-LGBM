```python
import pandas as pd
import numpy as np
import Benchmarks
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
import scipy.stats as stats
import time
# ID: 0RGI,1BBOB,2Affine,3Zigzag,4GPB,5RGF,6MA-BBOB
```

```python
# AS-LGBM + BBOB
np.random.seed(42)
ela,label=Benchmarks.get_dataset(1)
(TrainX, TestX,label1, label2) = train_test_split(ela, label, test_size=0.2,random_state=42)
TrainY=label1[:,0]
TrainR=label1[:,1:]
TestY=label2[:,0]
TestR=label2[:,1:]

train_data = lgb.Dataset(TrainX, label=TrainY)
test_data = lgb.Dataset(TestX, label=TestY)
evals_result = {}
params = {
        'objective': 'multiclass',
        'metric': 'multi_logloss',
        'learning_rate': 0.001,
        'max_depth': 5,
        'num_class': 10,
        # 'lambda_l1':0.1,
        # 'lambda_l2':0.1,
        "early_stopping_round": 200,
        "verbose_eval": 10,
        'verbosity': -1,
        # 'num_leaves':10,
        # "device_type": 'GPU'
    }
num_round = 1000  # 迭代次数
callbacks= [lgb.log_evaluation(period=1), ]

t1=time.time()
model = lgb.train(params, train_data, num_round,valid_sets=[train_data,test_data],callbacks=callbacks)
t2=time.time()

y_pred = model.predict(TestX)
y_hat = np.argmax(y_pred,axis=1)

# accuracy = accuracy_score(TestY, y_hat)
# print("准确率:", accuracy)
# print(classification_report(TestY,y_hat))

acc=0
for i in range(len(TestY)):
    if TestY[i]==y_hat[i]:
        acc+=1
    elif stats.ranksums(TestR[i,int(y_hat[i]*30):int((y_hat[i]+1)*30)],TestR[i,int(TestY[i]*30):int((TestY[i]+1)*30)])[1] > 0.05:
        acc+=1
print("可接受率",acc/len(TestY))
print('time cost',t2-t1)
```

```python
model.save_model('lightgbm_model_BBOB.bin', num_iteration=model.best_iteration)
```
